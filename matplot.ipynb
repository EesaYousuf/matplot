```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.arange(1,11,1)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.plot(x,x**2)
plt.title('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.plot(x,x**2,c='red',label='x^2')
plt.plot(x,x**3,c='blue',label='x^3')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.show()
```

```python
# histograms
students=pd.read_csv("StudentsPerformance.csv")
students
# plt.hist()
plt.style.use('ggplot')
plt.hist(students['math score'],bins=50)
# plt.grid(True)
plt.show()
```

```python
# scatter plots
x=students['math score']
y=students['reading score']
z= students['writing score']
plt.Figure(figsize=(10,8))
plt.scatter(x,y,c=z,cmap='Accent')
plt.colorbar(label='writing score')
plt.xlabel('math score')
plt.ylabel('Reading score')
plt.show()
```
